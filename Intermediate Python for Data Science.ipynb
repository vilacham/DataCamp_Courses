In the code cell below, we'll start working on the data that professor Hans Rosling used to build his beatiful bubble chart. It was collected in 2007. We'll work with two lists:
* `life_expectancy`, which contains the life expectancy for each country
* `gdp_cap`, which contains the GDP per capita for each country expressed in USD

GDP stands for Gross Domestic Product. It basically represents the size of the economy of a country. If we divide it by the population we get the GDP per capita.

```python
# Import pandas
import pandas as pd

# Import dataset as a DataFrame and create the two lists described before
dataframe = pd.read_csv('gapminder.csv', index_col = 0)
life_expectancy = list(dataframe.life_exp)
gdp_cap = list(dataframe.gdp_cap)

# Clear the current figure
plt.clf()

# Make a line plot with gdp_cap on the x-axis and life_expectancy on the y-axis
plt.plot(gdp_cap, life_expectancy)

# Display the plot
plt.show()
```

## Scatter plot
When we have a time scale along the horizontal axis, the line plot is our friend. But in many other cases, when we're trying to assess if there is a correlation between two variables, for example, the scatter plot is the better choice. Below is an example of how to build a scatter plot:

```python
plt.scatter(x, y)
plt.show()
```

Again, the cell couldn't execute the code because there is no x nor y defined. In the next code cell, we'll continue working with life expectancy and GDP per capita for different countries in 2007, but this time using a scatter plot to visualize it. In order to make the correlation clearer, we'll also display the GDP per capita on a logarithmic scale.

```python
# Clear the current figure
plt.clf()

# Make a scatter plot with gdp_cap in the x-axis and life_expectancy on the y-axis
plt.scatter(gdp_cap, life_expectancy)

# Put the horizontal axis on a logarithmic scale
plt.xscale('log')

# Display the plot
plt.show()
```

As we can see above, the higher GDP usually corresponds to a higher life expectancy, or, in other words, there is a positive correlation.

In the code cell below, we'll try to determine if there is also a relation between life expectancy and population of a country. We will extract a list containing the population of the countries and continue using the list `life_expectancy`.

```python
# Extract the population from the DataFrame
population = list(dataframe['population'] / 1000000)

# Clear the current figure
plt.clf()

# Make a scatter plot with population in the x-axis and life_expectancy in the y-axis
plt.scatter(population, life_expectancy)

# Display the plot
plt.show()
```

As we can see, there is no clear relationship between population and life expectancy.
