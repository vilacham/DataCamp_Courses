## Dictionary to DataFrame
Pandas is an open-source library, providing high-performance, easy-to-use structures and data analysis tools for Python.

The DataFrame is one of Pandas' most important data structures. It's basically a way to store tabular data where you can label the rows and the columns. One way to build a DataFrame is from a dictionary.

In the next code cell, we'll be working with vehicle data from different countries. Each observation corresponds to a country and the column give information about the number of vehicles per capita, whether people drive left or right, and so on.

We will define three lists:

* `names`, containing the country names for which data is available
* `dr`, a list with booleans that tells whether people drive left or right in the corresponding country
* `cpc`, the number of motor vehicles per 1000 people in the corresponding country

Each dictionary key will be a column label and each value will be a list which contains the column elements.

```python
# Import pandas
import pandas as pd

# Create lists
names = ['United States', 'Australia', 'Japan', 'India', 'Russia', 'Morocco', 'Egypt']
dr =  [True, False, False, False, True, True, True]
cpc = [809, 731, 588, 18, 200, 70, 45]

# Create dictionary with three key-value pairs
dictionary = {'country': names, 
              'drives_right': dr, 
              'cars_per_capita': cpc}

# Build a dataframe from dictionary
cars = pd.DataFrame(dictionary)

# Print dataframe
print(cars)
```

Since we don't want row labels as integers, we will create a list and use it to specify the row labels of the `cars` DataFrame.

```python
# Create list
row_labels_list = ['US', 'AUS', 'JAP', 'IN', 'RU', 'MOR', 'EG']

# Specify row labels of the DataFrame
cars.index = row_labels_list

# Print dataframe
print(cars)
```

## Square brackets
We can index and select Pandas DataFrames in many different ways. The simplest, but not the most powerful way, is to use square brackets.

In the code cell below, we'll import the same cars data from a CSV file as a Pandas DataFrame. To select only the `cars_per_cap` column we can use:
* `cars['cars_per_cap']`, which gives a Pandas Series
* `cars[['cars_per_cap']]`, which gives a Pandas DataFrame

```python
# Import pandas
import pandas as pd

# Import data
cars = pd.read_csv('cars.csv', index_col=0)

# Print out country column as Pandas Series
print(cars['country'])

# Print out country as Pandas DataFrame
print(cars[['country']])

# Print out DataFrame containing country and drives_right columns
print(cars[['country', 'drives_right']])
```

Square brackets can do more than just selecting columns. We can also se them to get rows, or observations, of a DataFrame. The following call selects the first five rows from the `cars` DataFrame: `cars[0:5]`. The result is another DataFrame containing only the rows we specified.

Notice that we can only select rows using sqaure brackets if we specify a slice, like `0:5`. Also, we're using the integer indexes of the rows here, not the row labels!

```python
# Import pandas
import pandas as pd

# Import data
cars = pd.read_csv('cars.csv', index_col=0)

# Print out the first 3 observations
print(cars[0:3])

# Print out the fourth, fifth and sixth observations
print(cars[3:6])
```

## loc and iloc
With `loc` and `iloc` we can do practically any data selection operation on DataFrames we can think of. `loc` is label-based, which means that we have to specify rows and columns based on their row and column labels. `iloc` is integer index based, so we have to specify rows and columns by their integer index like we did in the previous code cell.

Below, we'll expore both `loc` and `iloc`:

```python
# Import pandas
import pandas as pd

# Import data
cars = pd.read_csv('cars.csv', index_col=0)

# Print out observation for Japan with loc
print(cars.loc['JAP'])

# Print out observation for Japan with iloc
print(cars.iloc[2])

# Print observations for Australia and Egypt with loc
print(cars.loc[['AUS', 'EG']])

# Print observations for Australia and Egypt with iloc
print(cars.iloc[[1, 6]])
```

`loc` and `iloc` also allow us to select both rows and columns from a DataFrame:

```python
# Import pandas
import pandas as pd

# Import data
cars = pd.read_csv("cars.csv", index_col=0)

# Print out drives_right value of Morocco with loc
print(cars.loc['MOR', 'drives_right'])

# Print out drives_right value of Morocco with iloc
print(cars.iloc[5, 2])

# Print sub-DataFrame with loc
print(cars.loc[['RU', 'MOR'], ['country', 'drives_right']])

# Print sub-DataFrame with iloc
print(cars.iloc[[4, 5], [1, 2]])
```

It's also possible to select only columns with `loc` and `iloc`. In both cases, we simply put a slice going from beginning to end in front of the comma:

```python
# Import pandas
import pandas as pd

# Import data
cars = pd.read_csv('cars.csv', index_col=0)

# Print out drives_right column as Series
print(cars.loc[:, 'drives_right'])

# Print out drives_right column as DataFrame
print(cars.loc[:, ['drives_right']])

# Print out cars_per_cap and drives_right as DataFrame
print(cars.loc[:, ['cars_per_cap', 'drives_right']])
```

# Logic, Control Flow and Filtering
Boolean logic is the foundation of decision-making in any Python program. In the next cells, we'll learn about different comparison operators, how we can combine them with boolean operators and how to use the boolean outcomes in control structures. We'll also learn to filter data from Pandas DataFrames using logic.

## Comparison Operators
To check if two Python values, or variables, are equal we can use `==`. To check for inequality, we need `!=`.

```python
# Comparison of booleans
print(True == False)

# Comparison of integers
print(-5 * 15 != 75)

# Comparison of strings
print("pyscript" == "PyScript")

# Compare a boolean with an integer
print(True == 1)
```

We can also compare using the less than and the greater than signs, `<` and `>`, respectively. We can combine them with and equals sign to get greater than or equal to and less than or equals to (`<=` and `>=`, respectively).

At this point, we need to pay attention with two things:
* `<=` and `>=` are valid syntax, but `=<` and `=>` are not
* For string comparison, Python determines the relationship based on alphabetical order

```python
# Comparison of integers
x = -3 * 6
print(x >= -10)

# Comparison of strings
y = "test"
print("test" <= y)

# Comparison of booleans
print(True > False)
```

Out of the box, we can also use comparison operators with Numpy arrays. In the next code cell, we'll create two Numpy arrays, `my_house` and `your_house`, and they both contain the areas for the kitchen, living room, bedroom and bathroom in the same order, so we can compare them.

```python
# Import numpy
import numpy as np

# Create arrays
my_house = np.array([18.0, 20.0, 10.75, 9.50])
your_house = np.array([14.0, 24.0, 14.25, 9.0])

# Which areas in my_house are greater than or equal to 18
print(my_house >= 18)

# Which area in my_house is smaller than your_house
print(my_house < your_house)
```

## Boolean operators
A boolean is either `1` or `0`, `True` or `False`. With boolean operators such as `and`, `or` and `not` we can combine these booleans to perform more advanced queries on our data.

In the code cell below, we'll define two variables representing areas: `my_kitchen` and `your_kitchen`. After that, we'll explore these booleans operators.

```python
# Define variables
my_kitchen = 18.0
your_kitchen = 14.0

# Check whether my_kitchen is grater than 10 and less than 18
print(my_kitchen > 10 and my_kitchen < 18)

# Check whether my_kitchen is less than 14 or greater than 17
print(my_kitchen < 14 or my_kitchen > 17)

# Check whether double of my_kitchen is less than triple of your_kitchen
print(my_kitchen * 2 < your_kitchen * 3)
```

## Boolean operators with NumPy


Before, the operational operators like `<` and `>=` worked with NumPy arrays out of the box. Unfortunately, this is not true for the boolean operators `and`, `or` and `not`. To use these operators with NumPy, we'll need `np.logical_and()`, `np.logical_or()` and `np.logical_not()`.

```python
# Create arrays
my_house = np.array([18.0, 20.0, 10.75, 9.50])
your_house = np.array([14.0, 24.0, 14.25, 9.0])

# Items in my_house greater than 18.5 or smaller than 10
print(np.logical_or(my_house > 18.5, my_house < 10))

# Both my_house and your_house smaller than 11
print(np.logical_and(my_house < 11, your_house < 11))
```

## while
The while loop is like a repeated if statement. The code is executed over and over again, as long as the condition is `True`.

```python
# while example
x = 1
while x < 5:
    print(x)
    x += 1
```

## for
In the code cell below, we will write a example code with a `for` loop that iterates over all elements of a list and prints out every element separately:

```python
# Create list
areas = [11.25, 18.0, 20.0, 10.75, 9.50]

# for loop
for area in areas :
    print(area)
```

Using a `for` loop to iterate over a list only gives us access to every list element in each run, one after the other. If we also want to access the index information, so where the list element we're iterating over is located, we can use `enumerate()`:

```python
# Change the for loop to use enumerate()
for index, area in enumerate(areas):
    print('Room {}: {}'.format(index, area))
```

For non-programmer folks, `Room 0: 11.25` is strange. It would be better if the count started at 1:

```python
# Change the code to make the output easier to read
for index, area in enumerate(areas):
    print('Room {}: {}'.format(index + 1, area))
```

## Loop over a NumPy array
If we're dealing with a 1D NumPy array, looping over all elements can be as simple as:

`for x in array:`

If we're dealing with a 2D Numpy array, it's more complicated. A 2D array is built up of multiple 1D arrays. To explicitly iterate over all separate elements of a multi-dimensional array, we'll need this syntax:

`for x in np.nditer(array):`

In the next code cell, we will work with the `baseball.csv` file to understand how to loop over NumPy arrays.

```python
# Import dataset
baseball = pd.read_csv('baseball.csv')[['Height', 'Weight']].as_matrix().tolist()

# Create arrays
np_baseball = np.array(baseball)
np_height = np_baseball[:, 0]

# For loop over np_height
for height in np_height :
    print('{} inches'.format(height))

# For loop over np_baseball
for i in np.nditer(np_baseball) :
    print(i)
```

## Random numbers

Randomness has many uses in science, art, statistics, cryptography, gaming, gambling, and other fields. We're going to use randomness to simulate a game.

All the functionality we need is contained in the `random` package, a sub-package of `numpy`. In the code cell below, we'll be using two functions from this package:
* `seed()`: sets a random seed, so that our results are the reproducible between simulations. As an argument, it takes an integer of our choosing. If we call the function, no output is generated.
* `rand()`: if we don't specify any arguments, it generates a random float between zero and one.

```python
# Import numpy
import numpy as np

# Set the seed
np.random.seed(123)

# Generate and print a random float
print(np.random.rand())
```

We can also use `randint()`, also a function of the `random` package, to generate integers randomly. In the code cell below, we'll roll a dice:

```python
# Import numpy
import numpy as np

# Set the seed
np.random.seed(123)

# Roll the dice three times and print it
print(np.random.randint(1, 7))
print(np.random.randint(1, 7))
print(np.random.randint(1, 7))
```

Now it's time to play with the Empire State Building bet, in which our next move depends on the number of eyes we throw with the dice (it can be perfectly coded with an `if` - `elif` - `else` structure).

In the bet:
* if the dice is 1 or 2, we go one step down;
* if the dice is 3, 4 or 5, we go one step up;
* if the dice is 6, we throw the dice again and the number of eyes is the number of steps we go up.

```python
# Import numpy
import numpy as np

# Set the seed
np.random.seed(123)

# Set the starting step at 50
step = 50

# Create and roll the dice
dice = np.random.randint(1, 7)

# Construct the control structure
if dice <= 2:
    step -= 1
elif dice <= 5:
    step += 1
else:
    step += np.random.randint(1, 7)

# Print out dice and step
print(dice)
print(step)
```

Below, we will use a `for` loop to create a random ralk in the Empire State Building:

```python
# Import numpy
import numpy as np

# Set the seed
np.random.seed(123)

# Initialize random walk
random_walk = [0]

# Create the for loop
for i in range(100):
    # Set step as the last element in random_ralk list
    step = random_walk[-1]
    
    # Roll the dice
    dice = np.random.randint(1, 7)
    
    # Determine next step
    if dice <= 2:
        step -= 1
    elif dice <= 5:
        step += 1
    else:
        step += np.random.randint(1, 7)
    
    # Updata random_walk list
    random_walk.append(step)

# Print random_walk list
print(random_walk)
```

If we take a closer look at the list above, we'll see that there's number below 0, which can't happen. Let's fix it!

```python
# Import numpy
import numpy as np

# Set the seed
np.random.seed(123)

# Initialize random walk
random_walk = [0]

# Create the for loop
for i in range(100):
    # Set step as the last element in random_ralk list
    step = random_walk[-1]
    
    # Roll the dice
    dice = np.random.randint(1, 7)
    
    # Determine next step
    if dice <= 2:
        step = max(0, step - 1) # We use max to make sure step is always above 0
    elif dice <= 5:
        step += 1
    else:
        step += np.random.randint(1, 7)
    
    # Updata random_walk list
    random_walk.append(step)

# Print random_walk list
print(random_walk)
```

In the next code cell, we will visualize the random walk:

```python
# Import pyplot
import matplotlib.pyplot as plt

# Plot random_walk
plt.plot(random_walk)

# Show the plot
plt.show()
```

A single random walk is one thing, but that doesn't tell us if we have a good chance at winning the bet.

To get an idea about how big our chances are of reaching 60 steps, we can repeatedly simulate the random walk and collect the results. That's exactly what we'll do below.

```python
# Import numpy
import numpy as np

# Set the seed
np.random.seed(123)

# Create list of random_walks
random_walks = []

# Simulate 10 random walks
for i in range(10):
    # Initialize random walk
    random_walk = [0]

    # Create the for loop
    for i in range(100):
        # Set step as the last element in random_ralk list
        step = random_walk[-1]

        # Roll the dice
        dice = np.random.randint(1, 7)

        # Determine next step
        if dice <= 2:
            step = max(0, step - 1) # We use max to make sure step is always above 0
        elif dice <= 5:
            step += 1
        else:
            step += np.random.randint(1, 7)

        # Updata random_walk list
        random_walk.append(step)
    
    # Update random_walks list
    random_walks.append(random_walk)

# Print random_walks list
print(random_walks)
```

Notice that `random_walks` is a list of lists. If we convert it to a NumPy array, we ca start making interesting plots...

```python
# Convert random_walks to a numpy array
np_random_walks = np.array(random_walks)

# Plot and show all random walks
plt.plot(np_random_walks)
plt.show()
```

That's not exactly what we expected. Maybe it is better if we transpose the array:

```python
# Transpose array
np_random_walks_transposed = np.transpose(np_random_walks)

# Plor and show all random walks transposed
plt.plot(np_random_walks_transposed)
plt.show()
```

With this neatly written code of ours, changing the number of times the random walk should be simulated is super-easy. We simply update the `range()` function in the top-level `for` loop.

```python
# Import numpy and pyplot
import numpy as np
import matplotlib.pyplot as plt

# Set the seed
np.random.seed(123)

# Create list of random_walks
random_walks = []

# Simulate 10 random walks
for i in range(250):
    # Initialize random walk
    random_walk = [0]

    # Create the for loop
    for i in range(100):
        # Set step as the last element in random_ralk list
        step = random_walk[-1]

        # Roll the dice
        dice = np.random.randint(1, 7)

        # Determine next step
        if dice <= 2:
            step = max(0, step - 1) # We use max to make sure step is always above 0
        elif dice <= 5:
            step += 1
        else:
            step += np.random.randint(1, 7)

        # Updata random_walk list
        random_walk.append(step)
    
    # Update random_walks list
    random_walks.append(random_walk)

# Convert random_walks to a numpy array and transpose it
np_random_walks = np.array(random_walks)
np_random_walks_transposed = np.transpose(np_random_walks)

# Plot and show random_walks
plt.plot(np_random_walks_transposed)
plt.show()
```

There's still something we forgot! We're a bit clumsy and we have a 0.1% chance of falling down. That calls for another random number generation. Basically, we can generate a random float between 0 and 1. If this value is less than or equal to 0.001, we should reset step to 0.

```python
# Import numpy and pyplot
import numpy as np
import matplotlib.pyplot as plt

# Set the seed
np.random.seed(123)

# Create list of random_walks
random_walks = []

# Simulate 10 random walks
for i in range(250):
    # Initialize random walk
    random_walk = [0]

    # Create the for loop
    for i in range(100):
        # Set step as the last element in random_ralk list
        step = random_walk[-1]

        # Roll the dice
        dice = np.random.randint(1, 7)

        # Determine next step
        if dice <= 2:
            step = max(0, step - 1) # We use max to make sure step is always above 0
        elif dice <= 5:
            step += 1
        else:
            step += np.random.randint(1, 7)
        
        # Implement clumsiness
        chance_of_falling_down = np.random.rand()
        if chance_of_falling_down <= 0.001:
            step = 0

        # Updata random_walk list
        random_walk.append(step)
    
    # Update random_walks list
    random_walks.append(random_walk)

# Convert random_walks to a numpy array and transpose it
np_random_walks = np.array(random_walks)
np_random_walks_transposed = np.transpose(np_random_walks)

# Plot and show random_walks
plt.plot(np_random_walks_transposed)
plt.show()
```

All these fancy visualizations have put us on a sidetrack. We still have to solve the million-dollar problem: *what are the odds that you'll reach 60 steps high on the Empire State Building?*

Basically, we want to know about the end points of all the random walks we've simulated. These end points have a certain distribution that we can visualize with a histogram.

```python
# Import numpy and pyplot
import numpy as np
import matplotlib.pyplot as plt

# Set the seed
np.random.seed(123)

# Create list of random_walks
random_walks = []

# Simulate 10 random walks
for i in range(500):
    # Initialize random walk
    random_walk = [0]

    # Create the for loop
    for i in range(100):
        # Set step as the last element in random_ralk list
        step = random_walk[-1]

        # Roll the dice
        dice = np.random.randint(1, 7)

        # Determine next step
        if dice <= 2:
            step = max(0, step - 1) # We use max to make sure step is always above 0
        elif dice <= 5:
            step += 1
        else:
            step += np.random.randint(1, 7)
        
        # Implement clumsiness
        chance_of_falling_down = np.random.rand()
        if chance_of_falling_down <= 0.001:
            step = 0

        # Updata random_walk list
        random_walk.append(step)
    
    # Update random_walks list
    random_walks.append(random_walk)

# Convert random_walks to a numpy array and transpose it
np_random_walks = np.array(random_walks)
np_random_walks_transposed = np.transpose(np_random_walks)

# Select all the last rows from np_random_walks_transposed
ends = np_random_walks_transposed[-1]

# Plot and show random_walks
plt.hist(ends)
plt.show()
```

The histogram above was created from a Numpy array `ends`, that contains 500 integers. Each integer represents the end point of a random walk. To calculate the chance that this end point is greater than or equal to 60, we can count the number of integers in `ends` that are greater than or equal to 60 and divide that number by 500, the total number of simulations.

That said, in the code cell below we will calculate the estimated chance that we'll reach 60 steps high if we play this game.

```python
# Count how many times step is equal or greater than 60
count = [step for step in ends if step >= 60]

# Calculate probability
prob = (len(count) / len(ends)) * 100

# Print out the probability
print('Probability of reaching 60 steps or high: {:.1f}%.'.format(prob))
```
