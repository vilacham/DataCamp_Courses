```python
# Import pandas
import pandas as pd

# Import dataset as a DataFrame and create the two lists described before
dataframe = pd.read_csv('gapminder.csv', index_col = 0)
life_expectancy = list(dataframe.life_exp)
gdp_cap = list(dataframe.gdp_cap)

# Clear the current figure
plt.clf()

# Make a line plot with gdp_cap on the x-axis and life_expectancy on the y-axis
plt.plot(gdp_cap, life_expectancy)

# Display the plot
plt.show()
```

## Scatter plot
When we have a time scale along the horizontal axis, the line plot is our friend. But in many other cases, when we're trying to assess if there is a correlation between two variables, for example, the scatter plot is the better choice. Below is an example of how to build a scatter plot:

```python
plt.scatter(x, y)
plt.show()
```

Again, the cell couldn't execute the code because there is no x nor y defined. In the next code cell, we'll continue working with life expectancy and GDP per capita for different countries in 2007, but this time using a scatter plot to visualize it. In order to make the correlation clearer, we'll also display the GDP per capita on a logarithmic scale.

```python
# Clear the current figure
plt.clf()

# Make a scatter plot with gdp_cap in the x-axis and life_expectancy on the y-axis
plt.scatter(gdp_cap, life_expectancy)

# Put the horizontal axis on a logarithmic scale
plt.xscale('log')

# Display the plot
plt.show()
```

As we can see above, the higher GDP usually corresponds to a higher life expectancy, or, in other words, there is a positive correlation.

In the code cell below, we'll try to determine if there is also a relation between life expectancy and population of a country. We will extract a list containing the population of the countries and continue using the list `life_expectancy`.

```python
# Extract the population from the DataFrame
population = list(dataframe['population'] / 1000000)

# Clear the current figure
plt.clf()

# Make a scatter plot with population in the x-axis and life_expectancy in the y-axis
plt.scatter(population, life_expectancy)

# Display the plot
plt.show()
```

As we can see, there is no clear relationship between population and life expectancy.

## Histogram
To see how life expectancy in different countries is distributed, we will now create a histogram of `life_expectancy`. We won't specify the number of bins (Python will set the number of bins to 10 by default).

```python
# Clear the current figure
plt.clf()

# Create histogram of life_expectancy
plt.hist(life_expectancy)

# Display the plot
plt.show()
```

In the histogram above, we didn't specify the number of bins. By default, Python sets the number of buns to 10 in that case. The number of bins is pretty important. Too few bins will oversimplify reality and won't show us the details. Too many bins will overcomplicate reality and won't show the bugger picture. To control the number of bins to divide our data in, we can set the `bins` argument.

In the code cell below, we'll make two plots: one containing 5 bins and another containing 20 bins.

```python
# Clear the current figure
plt.clf()

# Create and show histogram of life_expectancy with 5 bins
plt.hist(life_expectancy, bins=5)
plt.show()

# Clear the current figure
plt.clf()

# Create and show histogram of life_expectancy with 20 bins
plt.hist(life_expectancy, bins=20)
plt.show()
```

## Customization
In this section, we're going to work on the scatter plot with world development data: GDP per capita in the x-axis (logarithmic scale), life expectancy in the y-axis.

```python
# Clear the current figure
plt.clf()

# Make scatter plot with logarithmic scale horizontal axis
plt.scatter(gdp_cap, life_expectancy)
plt.xscale('log')

# Display scatter plot
plt.show()
```

As a first step, we'll add axis labels and a title to the plot. We can do this with `xlabel()`, `ylabel()` and `title()` functions.

```python
# Clear the current figure
plt.clf()

# Make scatter plot with logarithmic scale horizontal axis
plt.scatter(gdp_cap, life_expectancy)
plt.xscale('log')

# Create variables
x_label = "GDP per capita (in USD)"
y_label = "Life expectancy (in years)"
title = "World Development in 2007"

# Add axis labels
plt.xlabel(x_label)
plt.ylabel(y_label)

# Add title
plt.title(title)

# Display customized scatter plot
plt.show()
```

In the next code cell, we'll add code to customize ticks. Our goal is to replace the values `1000`, `10000` and `100000` by `1k`, `10k` and `100k`, respectively.

```python
# Clear the current figure
plt.clf()

# Make scatter plot with logarithmic scale horizontal axis
plt.scatter(gdp_cap, life_expectancy)
plt.xscale('log')

# Create variables
x_label = "GDP per capita (in USD)"
y_label = "Life expectancy (in years)"
title = "World Development in 2007"
tick_values = [1000, 10000, 100000]
tick_labels = ['1k', '10k', '100k']

# Add axis labels
plt.xlabel(x_label)
plt.ylabel(y_label)

# Add title
plt.title(title)

# Customize ticks in the horizontal axis
plt.xticks(tick_values, tick_labels)

# Display customized scatter plot
plt.show()
```
