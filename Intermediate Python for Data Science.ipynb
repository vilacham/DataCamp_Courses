## Create a dictionary
In the code cell below, we'll create a dictionary named `europe`. As a refresher, here is a recipe for creating a dictionary:

dictionary = {'key1': 'value1', 'key2': 'value2'}

```python
# Create europe dictionary
europe = {'spain': 'madrid', 'france': 'paris', 'germany': 'berlin', 'norway': 'oslo'}

# Print europe
print(europe)
```

## Access dictionary
If the keys of a dictionary are chosen wisely, accessing its values is easy and intuitive. For example, to get the capital of France from `europe` we can use:

europe['france']

Here, `'france'` is the key and `'paris'` is the returned value. In the code cell below, we'll explore it a little bit more.

```python
# Print out the keys in europe dictionary
print(europe.keys())

# Print out value that belongs to key Norway
print(europe['norway'])
```

## Dictionary manipulation

```python
# Add Italy to the dictionary
europe['italy'] = 'rome'

# Check if Italy was properly added to the dictioinary
print('italy' in europe)

# Add Poland to the dictionary
europe['poland'] = 'warsaw'

# Print the dictionary
print(europe)
```

## Dictionary to DataFrame
Pandas is an open-source library, providing high-performance, easy-to-use structures and data analysis tools for Python.

The DataFrame is one of Pandas' most important data structures. It's basically a way to store tabular data where you can label the rows and the columns. One way to build a DataFrame is from a dictionary.

In the next code cell, we'll be working with vehicle data from different countries. Each observation corresponds to a country and the column give information about the number of vehicles per capita, whether people drive left or right, and so on.

We will define three lists:

* `names`, containing the country names for which data is available
* `dr`, a list with booleans that tells whether people drive left or right in the corresponding country
* `cpc`, the number of motor vehicles per 1000 people in the corresponding country

Each dictionary key will be a column label and each value will be a list which contains the column elements.

```python
# Import pandas
import pandas as pd

# Create lists
names = ['United States', 'Australia', 'Japan', 'India', 'Russia', 'Morocco', 'Egypt']
dr =  [True, False, False, False, True, True, True]
cpc = [809, 731, 588, 18, 200, 70, 45]

# Create dictionary with three key-value pairs
dictionary = {'country': names, 
              'drives_right': dr, 
              'cars_per_capita': cpc}

# Build a dataframe from dictionary
cars = pd.DataFrame(dictionary)

# Print dataframe
print(cars)
```

Since we don't want row labels as integers, we will create a list and use it to specify the row labels of the `cars` DataFrame.

```python
# Create list
row_labels_list = ['US', 'AUS', 'JAP', 'IN', 'RU', 'MOR', 'EG']

# Specify row labels of the DataFrame
cars.index = row_labels_list

# Print dataframe
print(cars)
```

## Square brackets
We can index and select Pandas DataFrames in many different ways. The simplest, but not the most powerful way, is to use square brackets.

In the code cell below, we'll import the same cars data from a CSV file as a Pandas DataFrame. To select only the `cars_per_cap` column we can use:
* `cars['cars_per_cap']`, which gives a Pandas Series
* `cars[['cars_per_cap']]`, which gives a Pandas DataFrame

```python
# Import pandas
import pandas as pd

# Import data
cars = pd.read_csv('cars.csv', index_col=0)

# Print out country column as Pandas Series
print(cars['country'])

# Print out country as Pandas DataFrame
print(cars[['country']])

# Print out DataFrame containing country and drives_right columns
print(cars[['country', 'drives_right']])
```

Square brackets can do more than just selecting columns. We can also se them to get rows, or observations, of a DataFrame. The following call selects the first five rows from the `cars` DataFrame: `cars[0:5]`. The result is another DataFrame containing only the rows we specified.

Notice that we can only select rows using sqaure brackets if we specify a slice, like `0:5`. Also, we're using the integer indexes of the rows here, not the row labels!

```python
# Import pandas
import pandas as pd

# Import data
cars = pd.read_csv('cars.csv', index_col=0)

# Print out the first 3 observations
print(cars[0:3])

# Print out the fourth, fifth and sixth observations
print(cars[3:6])
```

## loc and iloc
With `loc` and `iloc` we can do practically any data selection operation on DataFrames we can think of. `loc` is label-based, which means that we have to specify rows and columns based on their row and column labels. `iloc` is integer index based, so we have to specify rows and columns by their integer index like we did in the previous code cell.

Below, we'll expore both `loc` and `iloc`:

```python
# Import pandas
import pandas as pd

# Import data
cars = pd.read_csv('cars.csv', index_col=0)

# Print out observation for Japan with loc
print(cars.loc['JAP'])

# Print out observation for Japan with iloc
print(cars.iloc[2])

# Print observations for Australia and Egypt with loc
print(cars.loc[['AUS', 'EG']])

# Print observations for Australia and Egypt with iloc
print(cars.iloc[[1, 6]])
```

`loc` and `iloc` also allow us to select both rows and columns from a DataFrame:

```python
# Import pandas
import pandas as pd

# Import data
cars = pd.read_csv("cars.csv", index_col=0)

# Print out drives_right value of Morocco with loc
print(cars.loc['MOR', 'drives_right'])

# Print out drives_right value of Morocco with iloc
print(cars.iloc[5, 2])

# Print sub-DataFrame with loc
print(cars.loc[['RU', 'MOR'], ['country', 'drives_right']])

# Print sub-DataFrame with iloc
print(cars.iloc[[4, 5], [1, 2]])
```

It's also possible to select only columns with `loc` and `iloc`. In both cases, we simply put a slice going from beginning to end in front of the comma:

```python
# Import pandas
import pandas as pd

# Import data
cars = pd.read_csv('cars.csv', index_col=0)

# Print out drives_right column as Series
print(cars.loc[:, 'drives_right'])

# Print out drives_right column as DataFrame
print(cars.loc[:, ['drives_right']])

# Print out cars_per_cap and drives_right as DataFrame
print(cars.loc[:, ['cars_per_cap', 'drives_right']])
```

What a drill on indexing and selecting data from Pandas DataFrames!

# Logic, Control Flow and Filtering
Boolean logic is the foundation of decision-making in any Python program. In the next cells, we'll learn about different comparison operators, how we can combine them with boolean operators and how to use the boolean outcomes in control structures. We'll also learn to filter data from Pandas DataFrames using logic.

## Comparison Operators
To check if two Python values, or variables, are equal we can use `==`. To check for inequality, we need `!=`.

```python
# Comparison of booleans
print(True == False)

# Comparison of integers
print(-5 * 15 != 75)

# Comparison of strings
print("pyscript" == "PyScript")

# Compare a boolean with an integer
print(True == 1)
```

We can also compare using the less than and the greater than signs, `<` and `>`, respectively. We can combine them with and equals sign to get greater than or equal to and less than or equals to (`<=` and `>=`, respectively).

At this point, we need to pay attention with two things:
* `<=` and `>=` are valid syntax, but `=<` and `=>` are not
* For string comparison, Python determines the relationship based on alphabetical order

```python
# Comparison of integers
x = -3 * 6
print(x >= -10)

# Comparison of strings
y = "test"
print("test" <= y)

# Comparison of booleans
print(True > False)
```

Out of the box, we can also use comparison operators with Numpy arrays. In the next code cell, we'll create two Numpy arrays, `my_house` and `your_house`, and they both contain the areas for the kitchen, living room, bedroom and bathroom in the same order, so we can compare them.

```python
# Import numpy
import numpy as np

# Create arrays
my_house = np.array([18.0, 20.0, 10.75, 9.50])
your_house = np.array([14.0, 24.0, 14.25, 9.0])

# Which areas in my_house are greater than or equal to 18
print(my_house >= 18)

# Which area in my_house is smaller than your_house
print(my_house < your_house)
```

## Boolean operators
A boolean is either `1` or `0`, `True` or `False`. With boolean operators such as `and`, `or` and `not` we can combine these booleans to perform more advanced queries on our data.

In the code cell below, we'll define two variables representing areas: `my_kitchen` and `your_kitchen`. After that, we'll explore these booleans operators.

```python
# Define variables
my_kitchen = 18.0
your_kitchen = 14.0

# Check whether my_kitchen is grater than 10 and less than 18
print(my_kitchen > 10 and my_kitchen < 18)

# Check whether my_kitchen is less than 14 or greater than 17
print(my_kitchen < 14 or my_kitchen > 17)

# Check whether double of my_kitchen is less than triple of your_kitchen
print(my_kitchen * 2 < your_kitchen * 3)
```

## Boolean operators with NumPy
